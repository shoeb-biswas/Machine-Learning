* number of line in the string
* number of words in the string
* number of characters in the string
* And save it to the counter_of_string.txt file

```python
from functools import reduce
with open("./sample_data/test.txt","r") as file:

  #Number of lines
  string = file.readlines()
  line = len(string)

  #Number of word
  words = list(map(lambda x: len(x.split()),string))
  words = reduce(lambda x,y: x+y,words)

  #Number of character
  string = list(map(str.strip,string))
  string = list(map(lambda x: x.replace(" ",""),string))

  character = list(map(lambda x: len(x),string))
  character = reduce(lambda x,y: x+y, character)

  with open("./sample_data/counter_of_string.txt","w") as file:
    file.write(f"Total number of lines: {line}\nTotal number of words: {words}\nTotal number of character: {character}")
```

# **OOP (Class & Object)**

```python
# Class
class Phone:
  category = "Electronics"

  # Constructor
  def __init__(self, battry, camera, model, battry_level=100):
    self.battry = battry
    self.camera = camera
    self.model = model
    self.battry_level = battry_level
# Method
  def charge(self,hour):
    print(f"Charge completed by {hour}")
  def capture(self,photo):
    if (self.battry_level)<=0:
      print("No Charge")
    else:
      self.battry_level-=photo
      print(f"Captured on {self.model}")
```

```python
class cooling_mechanism:
  def __init__(self,cooling_method):
    self.cooling_method = cooling_method
  def cooling_on(self):
    print(f"The system is being cool by {self.cooling_method}")
```

```python
apple = Phone("3000mah", "12MP", "Iphone17")
apple.category = "Technology"
blueberry = Phone("5000mah","200MP","Ibot")
vivo = Phone("7000mah","20MP","Vivo21")

print(apple.camera)
print(blueberry.battry)
print(vivo.model)
apple.capture(10)
print(apple.battry_level)
```

```python
# Inharitance
class smartphone(Phone):
  def __init__(self, battry, camera, model, processor):
    super().__init__(battry, camera, model)
    self.processor = processor
  def charge(self,hour):
    print("Fast charging in process")
    super().charge(hour)
```

```python
pro = smartphone(120,23,"X", "Snapdragon")
print(pro.battry_level)

pro.charge(1)
```

# Multipe Inharitance

```python
class smartphone_cooling_mode(smartphone,cooling_mechanism):
  def __init__(self, battry, camera, model, processor, cooling_method):
    smartphone.__init__(self,battry, camera, model, processor)
    cooling_mechanism.__init__(self, cooling_method)
```

```python
pro_cooling = smartphone_cooling_mode("300mAh","72MP","Safari","Apple Bionic","Lequid Nitrogen")

print(pro_cooling.battry) # from phone class
print(pro_cooling.processor) # from smartphone class
print(pro_cooling.cooling_method) # from cooling_mechanism class
print(pro_cooling.cooling_on()) # Method of cooling_mechanism
print(pro_cooling.charge(20)) # Smartphone's modified charge inharited from phone class
```

# **Polymorphism**

```python
class camera:
  def __init__(self,name):
    self.name = name

# method
  def capture(self):
    print("A photo is captured")
```

```python
class Smart(camera):
  def __init__(self,name,resolution):
    super().__init__(name)
    self.resolution = resolution

# Method Overriding
  def capture(self):
    print("Photo is captured by Phone")


class DSLR(camera):
  def __init__(self,name,resolution):
    super().__init__(name)
    self.resolution = resolution

# Method Overriding
  def capture(self):
    print("Photo is captured by DSLR")

class Drone(camera):
  def __init__(self,name,resolution):
    super().__init__(name)
    self.resolution = resolution

# Method Overriding
  def capture(self):
    print("Photo is captured by Drone")

phone = Smart("Iphone","48MP")
dslr = DSLR("Nicon","4K")
drone = Drone("Amazon","1080p")

phone.capture()
dslr.capture()
drone.capture()
```

# **Encapsulation**

```python
class mobile:
  def __init__(self,name,model,imei):
    self.__name = name
    self.__model = model
    self.__imei = imei #private

  def charge(self):
    print("Phone is charging")

  # To access IMEI
  def imei_gatter(self):
    return self.__imei

  def model_gatter(self):
    return self.__model

  def name_gatter(self):
    return self.__name

  def name_setter(self,name):
    self.__name = name


iphone = mobile("Phone",17,"1x343")

print(iphone.name_setter("Iphone"))
print(iphone.model_gatter())
print(iphone.imei_gatter())
print(iphone.name_gatter())
```
