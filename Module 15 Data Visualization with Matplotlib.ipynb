# 2D Line from Pandas DF

```python
df = pd.read_csv('enrollment_data.csv')
```

```python
df
```

```python
plt.xlabel('Years')
plt.ylabel('Enrollment')
plt.title('Student enrollment over 8 years')

plt.plot(df['Year'],df['Programming'])
plt.plot(df['Year'],df['Digital Marketing'])
```

```python
plt.xlabel('Years')
plt.ylabel('Enrollment')
plt.title('Student enrollment over 8 years')

plt.plot(df['Year'],df['Programming'],label = 'Programming' )
plt.plot(df['Year'],df['Digital Marketing'],label = 'Digital Marketing')

plt.legend()
```

```python
plt.xlabel('Years')
plt.ylabel('Enrollment')
plt.title('Student enrollment over 8 years')

plt.plot(df['Year'],df['Programming'],label = 'Programming', color = '#b042f5', linewidth = 5, linestyle='dashed',marker ='o',markersize=7)
plt.plot(df['Year'],df['Digital Marketing'],label = 'Digital Marketing',color = 'black', linewidth = 3,linestyle='dotted')

plt.legend(loc='lower right')
plt.grid()
plt.show()
```

# Bar Chart

```python
data = {
    'Student': ['B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'StudyHours': [ 2, 5, 3, 7, 4, 8, 1],
    'Attendance': [ 60, 88, 70, 95, 80, 98, 55],
    'Grade': [ 'Fail', 'Pass', 'Fail', 'Pass', 'Pass', 'Pass', 'Fail']
}

df = pd.DataFrame(data)
```

```python
df
```

```python
# Gender wise catagorise
Gender_group = df.groupby('Gender').size()
Gender_group.index
Gender_group.values
```

```python
plt.bar(Gender_group.index,Gender_group.values, color= ['red','orange'],edgecolor = 'black')

plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Bar chart of gender and count')
```

```python
# Bar chart on real data
df1 = pd.read_csv('student_data.csv')
df1
```

```python
status = df1.groupby('CompletionStatus').size()

plt.bar(status.index, status.values, color= ['green','black','red'], edgecolor = 'yellow' )

plt.xlabel('CompletionStatus')
plt.ylabel('Count')
plt.title('Bar chart of Completion Status')
```

```python
status = df1.groupby('Location').size()

plt.bar(status.index, status.values,color= 'green' ,edgecolor = 'red' )

plt.xlabel('Location')
plt.ylabel('Count')
plt.title('Bar chart of Location')
```

# Pie Chart

```python
status = df1.groupby('Location').size()
plt.pie(status, labels= status.index,autopct='%1.1f%%')
plt.show()
status
```

```python
status = df1.groupby('CompletionStatus').size()
plt.pie(status, labels= status.index,autopct='%1.1f%%', explode=(0.09,0,0),shadow= True)
plt.show()
status
```
