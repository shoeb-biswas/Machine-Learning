Create a text file named emails.txt containing multiple email addresses, some in uppercase and some repeated.

Problem 1: Common Friends (Set Operations)
Problem:
 Given two sets of friends from two people, find the mutual friends, unique friends of each, and the total number of unique friends.
Example Input:
a_friends = {"Rahim", "Karim", "Sakib", "Jamal"}
b_friends = {"Sakib", "Jamal", "Rafiq", "Nadim"}

Expected Output:
Mutual friends: {'Sakib', 'Jamal'}
Unique to A: {'Rahim', 'Karim'}
Unique to B: {'Rafiq', 'Nadim'}
Total unique friends: 6


```python
a_friends = input().split()
b_friends = input().split()

# Mutual
uniqe1 = set(a_friends)
uniqe2 = set(b_friends)

mutual = uniqe1.intersection(uniqe2)

intera = uniqe1.difference(uniqe2)
interb = uniqe2.difference(uniqe1)
total = len(intera) + len(interb)

print(f"Mutual friends: {mutual}")
print(f"Unique to A: {intera}")
print(f"Unique to B: {interb}")
print(f"Total unique friends: {total}")



```

# Problem 2: Sort Students by Marks (Tuple + Dictionary)
* **Problem:**
 You are given a list of tuples ‚Äî each containing a student‚Äôs name and mark.
 Sort them in descending order of marks and print the top 3 scorers.
* **Example Input:**
* students = [("Rafi", 89), ("Sumi", 95), ("Hasan", 90), ("Nila", 75), ("Anik", 98)]

* **Expected Output:**
* Top 3 students:
* Anik - 98
* Sumi - 95
* Hasan - 90


```python
students = [("Rafi", 89), ("Sumi", 95), ("Hasan", 90), ("Nila", 75), ("Anik", 98)]
students = dict(students)
sorted_marks = sorted(students.items(), key=lambda x: x[1], reverse=True)
Top3 = sorted_marks[:3]

print("Top 3 Students:")
for name, mark in Top3:
  print(f"{name} - {mark}")
```

# Problem 3: Word Length Filter (Lambda + Filter)
* **Problem:**
 Given a list of words, use filter() and lambda to return only words whose length is greater than 4.
* **Example Input:**
words = ["sun", "planet", "moon", "star", "universe"]

* **Expected Output:**
['planet', 'universe']


```python
words = ["sun", "planet", "moon", "star", "universe"]
words_pro = list(filter(lambda x: len(x)>4,words))
print(words_pro)
```

# Problem 4: Square of Even Numbers (Map + Filter + Lambda)
* **Problem:**
 Write a one-line Python expression using map(), filter(), and lambda that takes a list of integers and returns the squares of even numbers only.
* **Example Input:**
nums = [1, 2, 3, 4, 5, 6]

* **Expected Output:**
[4, 16, 36]


```python
nums = [1, 2, 3, 4, 5, 6]
square = list(map(lambda x: x**2, filter(lambda x: x%2==0,nums)))
print(square)
```

1. Make three classes:
* Phone ‚Üí has model and battery (default 100).


* SmartPhone ‚Üí inherits from Phone and adds operating system.


* GamingPhone  ‚Üí inherits from SmartPhone and adds cooling_system.
 * Add a method start_game(name) that prints
 * üëâ "Playing <name> on <model>".

```python
class Phone:
  def __init__(self,model, battery=100):
    self.model = model
    self.battery = battery
```

```python
class SmartPhone(Phone):
  def __init__(self,model, battery,operatingsystem):
    super().__init__(model,battery)
    self.operatingsystem = operatingsystem
```

```python
class GamingPhone(SmartPhone):
  def __init__(self,model, battery,operatingsystem,cooling_system):
    super().__init__(model, battery,operatingsystem)
    self.cooling_system = cooling_system

  def start_game(self,name):
    print(f"Playing {name} on {self.model}")
```

```python
phone = GamingPhone("Iphone","85","ios","Vapping Pannel")
phone.start_game("Freefire")
```
