## 14.1 Filtering on string

```python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Grace", "Hannah","Saku"],
    "City": ["New York", "Los Angeles", "Newark", "Boston", "New Delhi", "Chicago", "New Orleans", "Houston","A Los Carol"],
    "Department": ["HR", "IT", "Finance", "IT", "HR", "Marketing", "Finance", "HR", "HR"],
    "Salary": [50000, 60000, 55000, 70000, 52000, 58000, 62000, 51000,1245]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
df.loc[df['City'].str.contains("New")]
```

```python
# To false down the case sensativeness
df.loc[df['City'].str.contains("new",case = False)]
```

```python
# Start with los [print all Los]
df.loc[df['City'].str.contains(r"Los")]
```

```python
# Print just start with Los
df.loc[df['City'].str.contains(r"^Los")]
```

```python
# Ending with rk
df.loc[df['City'].str.contains(r"rk$")]
```

```python
# Start with Vowel
df.loc[df['Name'].str.contains(r"^[AEIOU]")]
```

```python
# Include New or Los
df.loc[df['City'].str.contains(r"New|Los")]
```

# Apply Function in DF

```python
mn = df1['Total Marks'].min()
mx = df1['Total Marks'].max()
df1['Scaled Marks'] = df1['Total Marks'].apply(lambda x: (x-mn)/(mx-mn))
```

```python
# Custom made Function

def grading_system(marks):
    if marks >= 260:
        return 'A+'
    elif marks >= 250:
        return 'A'
    else:
        return 'A-'

df1['Grade'] = df1['Total Marks'].apply(grading_system)
    
```

```python
df1
```

```python
def Marking_System(df1):
    a = df1['Data Structure Marks'] * 3
    b = df1['Algorithm Marks'] * 2
    c = df1['Python Marks'] * 5
    return a+b+c


df1['Exceptional Marks'] = df1.apply(Marking_System,axis=1)
```

# Problem Solving 2

```python
import pandas as pd

def count_salary_categories(accounts: pd.DataFrame) -> pd.DataFrame:
    low_salary = (accounts['income']<20000).sum()
    average_salary = ((accounts['income']>=20000) & (accounts['income']<=50000)).sum()
    high_salary = (accounts['income']>50000).sum()

    data = [('Low Salary', low_salary),('Average Salary', average_salary),('High Salary', high_salary)]

    df = pd.DataFrame(data, columns = ['category','accounts_count'])
    return df
```
